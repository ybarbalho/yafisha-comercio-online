# Y.Afisha: Controle de despesas de marketing

## 1. Introdução

A Y.Afisha é uma loja de comércio on-line que deseja entender melhor suas campanhas de marketing e resultados, de modo a conhecer como funciona a atratividade do seu produto e quais os lucros obtidos sobre o trabalho dessas campanhas. Para a análise do nosso projeto, obtivemos os logs de acesso ao sistema durante o período de junho de 2017 a maio de 2018, assim como as compras realizadas no período e as despesas com marketing.

Dentre essas informações, temos a descrição das seguintes tabelas:

- Tabela **visits** (os logs do servidor com dados sobre os acessos ao site):

    *Uid:* identificador unívoco do usuário;
  
    *Device:* dispositivo do usuário;
  
    *Start Ts:* data e hora do início da sessão;
  
    *End Ts:* data e hora do final da sessão;
  
    *Source Id:* identificador da origem do anúncio através do qual o usuário chegou.

- Tabela **orders** (dados sobre os pedidos):

    *Uid:* identificador unívoco do usuário que faz um pedido;
  
    *Buy Ts:* data e hora do pedido;
  
    *Revenue:* a receita da Y.Afisha com o pedido.

- Tabela **costs** (dados sobre as despesas com marketing):

     *source_id:* identificador da origem de anúncio;
  
     *dt:* data;
  
     *costs:* despesas com esta origem de anúncio neste dia.

### 3.1. Analisando o padrão de visitas ao sistema

Primeiramente, para obtermos informações específicas sobre os acessos diários, semanais e mensais do nosso site, precisamos isolar colunas que contenham esses dados. Para isso, utilizaremos o método *dt.weekday* para transformar nossas datas em dias da semana, identificados de 0 a 6 (começando na segunda-feira), e o formato *datetime64* para trabalharmos datas semanais e mensais e suas diferenças em relação ao primeiro acesso.

```python
# Isolando as informações de datas referentes ao dia da semana, semana do ano e mês de cada acesso
df_visits['day_of_week'] = df_visits['start_ts'].dt.weekday
df_visits['week'] = df_visits['start_ts'].dt.to_period('W')
df_visits['month'] = df_visits['start_ts'].dt.to_period('M')
```

```python
# Observando as colunas criadas acima
df_visits.head()
```

Agora, criaremos coortes para identificar os grupos referentes às datas em estudo. Coortes são agrupamentos de dados referentes a eventos que aconteceram no mesmo período de tempo. No nosso caso, estudaremos os acessos (eventos) que aconteceram a cada dia da semana, semana do ano e mês (período de tempo).

Além disso, separaremos os agrupamentos pela contagem do número total de sessões e de acessos unívocos por usuário durante o período em estudo.

```python
# Criando as coortes de visitas por dia da semana, semana do ano e mês de acesso de cada sessão
cohort_dow = df_visits.groupby('day_of_week').agg({'uid': ['count', 'nunique']}) # count para cada sessão e nunique para cada usuário
cohort_week = df_visits.groupby('week').agg({'uid': ['count', 'nunique']})
cohort_month = df_visits.groupby('month').agg({'uid': ['count', 'nunique']})
```

```python
# Ajustando os nomes das colunas
cohort_dow.columns = ['n_sessions', 'n_users']
cohort_week.columns = ['n_sessions', 'n_users']
cohort_month.columns = ['n_sessions', 'n_users']
```

Com as coortes enfim criadas, seguimos para a análise gráfica dos nossos dados.

### 3.3. Duração dos acessos de usuário e duração média das sessões (ASL)

```python
# Criando o gráfico de distribuição da duração de cada sessão
df_visits['session_duration_sec'] = (df_visits['end_ts'] - df_visits['start_ts']).dt.seconds # Cálculo de duração da sessão (em segundos).

# Removendo sessões sem duração e transformando em minutos
(df_visits[df_visits['session_duration_sec']>0]['session_duration_sec']/60).hist(bins=100)
plt.xlabel('Duração da sessão (em minutos)')
plt.ylabel('Frequência')
plt.title('Distribuição de duração das sessões')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

O cálculo do tempo de duração de cada sessão do usuário em nosso site é feito através da subtração da data de saída do usuário do sistema pela data de acesso. Com essa informação obtida no formato de segundos, atribuímos uma nova coluna de duração *(session_duration_sec)* ao nosso dataframe.

No gráfico de distribuição acima, convertemos os valores para minutos e observamos que os usuários passam pouco tempo utilizando nosso sistema, sendo que a grande maioria deles não gasta nem 20 minutos consultando os produtos.

Abaixo iremos verificar a duração média das sessões (ASL), representativo real do tempo gasto por cada usuário ao longo do acesso.

```python
# Calculando a duração média da sessão (ASL) do usuário
asl = df_visits['session_duration_sec'].mean()
print(f'Duração média da sessão dos usuários (ASL): {asl/60:.2f} minutos')
```

Para o cálculo da **duração média das sessões de usuário (ASL)**, descobrimos a média da duração da sessão de todos os usuários do sistema. Como o cálculo feito previamente apresentava a duração em segundos, com a divisão do valor por 60 (segundos) obtemos a quantidade média de quantos minutos cada usuário gasta em nosso sistema.

O valor obtido de cerca de 10 minutos e meio mostra que os usuários não costumam passar muito tempo em nosso sistema. É possível que muitos deles já acessem o site tendo em mente o que vão comprar, e por isso não se interessam por novos produtos. Pode ser interessante que, em abordagens futuras do sistema, sejam apresentadas ofertas de compras para captar a atenção do usuário e aumentar o tempo de duração das próximas sessões.

### 3.4. Intervalo de tempo entre os acessos

Uma informação importante de se analisar é a de quanto tempo o usuário leva para retornar a acessar o sistema. Usuários recorrentes alavancam as receitas com a venda constante de produtos, e sua retenção deve ser uma das principais metas de futuras campanhas.

```python
# Criando uma coluna com a data do primeiro acesso de cada usuário e calculando quantos dias desde sua última visita
first_visits = df_visits.groupby('uid')['start_ts'].min()
df_visits = df_visits.join(first_visits.rename('first_visit'), on='uid')
df_visits['days_since_first'] = (df_visits['start_ts'] - df_visits['first_visit']).dt.days
```

```python
df_visits.head()
```

```python
df_visits['days_since_first'].value_counts().head(10)
```

Nos dados acima, podemos observar que tivemos mais de 244.000 primeiros acesso ao sistema. Como esse valor pode interferir na análise média do intervalo até a próxima visita, removeremos os primeiros dias de acesso do cálculo abaixo, para obter apenas a informação sobre visitas em dias seguintes.

```python
# Removemos 0 pois se trata do mesmo dia da primeira visita
df_visits_since_first = df_visits[df_visits['days_since_first']>0]['days_since_first']
```

```python
# Plotando a distribuição do intervalo de tempo entre acessos
df_visits_since_first.hist(bins=25, color='dodgerblue')
plt.xlabel('Nº de dias após o primeiro acesso')
plt.ylabel('Frequência')
plt.title('Distribuição do intervalo entre acessos')

# Delimitando o intervalo de exibição do eixo x para cada 15 dias
xticks = np.arange(0, df_visits['days_since_first'].max(), 15)
plt.xticks(xticks, rotation=90)

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Calculando a média de dias desde o último acesso do usuário
df_visits_since_first.mean()
```

No gráfico de distribuição do intervalo entre acessos desde o dia da primeira visita, podemos observar que a maior frequência se encontra no intervalo de até 15 dias de diferença desde a primeira visita. Isto é, os usuários retornam ao site num período de até 2 semanas após a sua primeira visita. Como a distribuição desse intervalo possui 125% mais ocorrências que a do próximo intervalo de tempo (até 30 dias), podemos apontar que a taxa de retorno de clientes ao nosso site é satisfatória.

Apesar disso, notamos que, em termos gerais, os usuários levam em média 93 dias, cerca de 3 meses, para retornar ao site. É um lapso temporal importante e que deve ser trabalhado para que tenhamos maior retenção de clientes nos próximos estudos, aumentando a frequência de acessos desses usuários menos recorrentes ao sistema e encurtando o intervalo de tempo em que realizam suas compras.

E-mails podem ser enviados ofertando produtos em promoção e campanhas publicitárias sazonais podem ser trabalhadas para períodos de baixa demanda.

### 3.7. Cálculo do Lifetime Value (LTV) dos clientes

O Lifetime Value (LTV) dos clientes é uma métrica que estima quanto dinheiro um cliente traz para a empresa durante o período do seu relacionamento com ela.

Essa métrica é importante para definir informações como quanto podemos investir para a captação de um cliente, a lucratividade do negócio e a taxa de retenção dos nossos clientes, auxiliando a tomada de decisões de marketing, produtos e vendas com base nesses dados.

```python
df_orders.head()
```

```python
df_orders.shape
```

```python
# Calculando o Lifetime Value (LTV) médio dos clientes
ltv = df_orders.groupby('uid')['revenue'].sum().mean()
print(f'Lifetime Value (LTV) médio dos clientes: ${ltv:.2f}')
```

O LTV médio no valor de $6,90 gerado pelos clientes ao longo do período em estudo representa, a princípio, bom resultado financeiro para a empresa.

Vamos aprofundar o estudo dessa métrica, verificando quanto de valor cada cliente gerou com base em grupos categorizados por mês da compra inicial.

```python
# Calculando o LTV gerado por cada cliente. Vamos gerar uma tabela com esses dados.
ltv_customers = df_orders.groupby('uid')['revenue'].sum().reset_index()

# Separando o primeiro mês de compra do cliente
first_month = df_orders.groupby('uid')['month'].min().reset_index()

# Realizando o merge das duas tabelas
ltv_customers = ltv_customers.merge(first_month, on='uid')

# Agrupamos e calculamos o LTV médio do grupo de clientes de cada mês
ltv_month = ltv_customers.groupby('month')['revenue'].mean().reset_index()
```

```python
ltv_month['month'] = ltv_month['month'].dt.strftime('%m-%Y')
```

```python
# Plotando o gráfico do LTV médio mensal pelo primeiro mês de compra
plt.figure(figsize=(12, 3))
plt.plot(ltv_month['month'], ltv_month['revenue'], marker='o', color='royalblue')

plt.title('LTV médio por mês da primeira compra')
plt.xlabel('')
plt.ylabel('LTV médio ($)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.show()
```

Podemos observar que os clientes que realizaram suas primeiras compras no mês de junho de 2017 se destacaram e geraram o segundo maior LTV para a empresa ao longo do período estudado. Os novos clientes de julho e agosto mantiveram o LTV na faixa dos $8.

Porém, é no mês de setembro que notamos o maior pico de clientes que geraram em média $14 de LTV para a empresa. A partir disso, os novos clientes não realizaram compras numerosas a ponto de obter destaque no gráfico, com exceção de um pequeno pico dos clientes de dezembro.

Dessa forma, reforçamos a necessidade de campanhas para promoção das vendas aos novos clientes, visando alavancar os números de vendas desse grupo.

```python
# Ajustando o nome da tabela do primeiro mês de compra
first_month = first_month.rename(columns={'month': 'first_month'})
```

```python
# Realizando o merge do primeiro mês na tabela de pedidos
df_orders = df_orders.merge(first_month, on='uid')

# Certificando o tipo de apresentação da coluna month (mês do pedido)
df_orders['month'] = df_orders['buy_ts'].dt.to_period('M')

# Criando uma coluna para a diferença entre o mês da atual compra e o primeiro mês
df_orders['age'] = df_orders['month'].astype('int64') - df_orders['first_month'].astype('int64')
```

```python
# Gerando os tamanhos de coorte para cada primeiro mês de compra
cohort_sizes = df_orders.groupby('first_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes = cohort_sizes.rename(columns={'uid': 'n_buyers'})
```

```python
# Separando as coortes por mês da primeira compra e somando sua receita
cohorts = df_orders.groupby(['first_month', 'age']).agg({'revenue': 'sum'}).reset_index()
```

```python
# Merge com o tamanho das coortes para calcular o LTV
report = cohorts.merge(cohort_sizes, on='first_month')

# Calculando o LTV acumulado para cada mês
report['ltv'] = report['revenue'] / report['n_buyers']
```

```python
# Gerando uma pivot table com os dados do LTV por coorte
result = report.pivot_table(
    index='first_month',
    columns='age',
    values='ltv',
    aggfunc='mean'
).cumsum(axis=1)
```

```python
# Plotando o gráfico de LTV por coorte em um heatmap
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.heatmap(result, annot=True, fmt='.3f', linewidths=1, linecolor='gray', cmap='YlGnBu')
plt.title("LTV por coorte")
plt.xlabel("Nº de meses desde a primeira compra")
plt.ylabel("Coorte (mês da primeira compra)")
plt.show()
```

Na separação por coortes do primeiro mês de compra de cada cliente, observamos que a receita gerada pelos clientes que fizeram a compra no 4º mês do estudo concentra a maior renda na janela de tempo observada. Em seguida, os clientes que realizam suas compras no primeiro mês de acesso também representam bons números no LTV gerado para a empresa, confirmando os dados observados na exibição do gráfico de linhas.

### 3.8. Cálculo do Custo de Aquisição de Cliente (CAC)

O Custo de Aquisição de Cliente (CAC) mede o custo médio de captação de um novo cliente. Essa é uma métrica importante para observarmos se a operação e os custos de marketing estão sendo lucrativos para a empresa. Para o negócio ser sustentável, é de comum senso que o valor do LTV seja ao menos o triplo do valor do CAC.

```python
# Calculando a soma da tabela de custos
total_costs = df_costs['costs'].sum()
print(f'Custo total com marketing: ${total_costs}')
```

```python
# Calculamos primeiramente o total de clientes únicos adquiridos no período.
distinct_customers = df_orders['uid'].nunique()
```

```python
# Após isso, com o valor do investimento total em mãos, calculamos o valor do CAC.
cac = total_costs/distinct_customers
print(f'Custo de Aquisição de Cliente (CAC): ${cac:.2f}')
```

O custo de aquisição de cada cliente no valor de $9,01 é maior que nosso LTV médio por cliente. Em suma, esses números demonstram que a empresa está gastando mais para atrair um cliente que o valor que esses clientes têm gerado de receita com seus pedidos.

Dessa forma, acendemos o alerta vermelho em relação aos gastos com marketing. A seguir vamos estudar seu retorno e analisar quais são os problemas.

### 3.11. Retorno sobre o investimento (ROI)

Chegou a hora de verificarmos se o nosso negócio é sustentável e lucrativo em relação aos seus gastos. Para tanto, iremos calcular o ROI, que nada mais é que o Retorno sobre o Investimento *(Return on Investment)*, um indicador financeiro que mede a eficiência de um investimento.

```python
# Retorno sobre o investimento (ROI) total
roi = (df_orders['revenue'].sum() - df_costs['costs'].sum()) / df_costs['costs'].sum()
print(f'Retorno sobre o investimento (ROI): ${roi:.2f}')
```

```python
total_costs = df_costs['costs'].sum()

# Cálculo do ROI percentual
print(f'ROI percentual: {((roi/total_costs) * 100):.4f}%')
```

Observamos que o valor do ROI e o seu percentual estão negativos, o que indica que os custos de marketing atualmente geram prejuízo à empresa. Os gastos estão muito altos e o retorno desse investimento esperado nas compras não está sendo atingido de forma produtiva.

Apesar do bom número de acessos proporcionado pelas campanhas de marketing, as visitas não estão sendo convertidas em clientes para nosso produto. Urge a necessidade do deslocamento nos investimentos com o marketing, ou a promoção de campanhas para captação dos clientes que visitam o site e não se tornam compradores.

Vamos observar esses dados nos gráficos a seguir.

```python
# Obtendo a receita gerada por origem de anúncio
revenue_by_source = df_orders.merge(first_source, on='uid').groupby('source_id')['revenue'].sum()

# Cálculo do ROI por origem de anúncio
roi_by_source = (revenue_by_source - cost_by_source) / cost_by_source
```

```python
# Fazendo a separação visual dos rendimentos positivos e negativos
colours = ['seagreen' if roi >= 0 else 'red' for roi in roi_by_source.dropna()]
```

```python
# Criando o gráfico para visualizar o ROI por origem de anúncio
roi_by_source.dropna().plot(kind='bar', color=colours)

plt.title('ROI por origem de anúncio')
plt.xlabel('Origem')
plt.ylabel('Valor do retorno ($)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

Como podemos observar acima, os anúncios da origem nº 1 tornaram essa a categoria de maior potencial lucrativo para a empresa, com base no valor do seu investimento original. O custo relativamente baixo para a captação de clientes por meio dessa categoria e a taxa de retorno de receita que eles geraram demonstra que devemos abrir os olhos e focar nessa forma de anúncio.

Além disso, as categorias de nº 2 e 5, que também não tiveram forte investimento, exibiram boa capacidade de atração de compradores para a empresa. Por outro lado, a categoria de anúncios nº 3, aquela que teve o maior investimento, não trouxe a quantidade prospectada de clientes compradores para a empresa, apresentando o maior prejuízo financeiro do período.

A categoria nº 9, mesmo sem grande investimento, apresentou o terceiro melhor resultado de retorno sobre o investimento feito. Sugerimos que esse canal de anúncios receba maior investimento para observarmos, em estudos futuros, a projeção de receita que ele pode gerar.

Por fim, a categoria de nº 10, que quase não teve investimento, gerou o segundo pior resultado para a companhia. Sendo assim, podemos abrir mão do investimento nessa categoria de anúncios, tendo em vista que o custo de aquisição de clientes é muito caro para, ao fim, gerar prejuízo.

Assim, sugerimos a revisão dos gastos nas categorias destacadas acima, principalmente o corte de investimento para os anúncios da categoria de origem nº 3.
